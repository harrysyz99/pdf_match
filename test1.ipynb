```python
import PyPDF2
import os

directory2 = "F:/Downloads/pdf_file"
file_list=[file for file in os.listdir(directory2) if file.endswith('.pdf')]
res=[]
for file in file_list:
    word_to_count1 = '政策'
    word_to_count2 = '风险'
    pdf_file = open(os.path.join(directory2,file), 'rb')
    pdf_reader = PyPDF2.PdfReader(pdf_file)

    num_pages = len(pdf_reader.pages)

    count1 = 0
    count2 = 0
    # 循环遍历所有页面
    for i in range(num_pages):
        page = pdf_reader.pages[i]

        text = page.extract_text()
        count1 += text.lower().count(word_to_count1.lower())
        count2 += text.lower().count(word_to_count2.lower())
    res1=True if count1 else False
    res2=True if count2 else False
    # print([res1,count1,res2,count2])
    res.append([res1,count1,res2,count2])
    import csv

    with open('result.csv','a',encoding='utf-8',newline='') as f:
        writer=csv.writer(f)
        writer.writerow([res1,count1,res2,count2])
    f.close()
```
